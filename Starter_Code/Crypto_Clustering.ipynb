```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
```

```python
# Create a DataFrame with the scaled data
# Set the coinid column as index

# Fit and transform the numerical data in the DataFrame
df_market_data.iloc[:, 1:] = scaler.fit_transform(df_market_data.iloc[:, 1:])

# Display sample data
df_market_data.head()


```

---

```python
# Copy the crypto names (index)
crypto_names = df_market_data.index.copy()

# Create a DataFrame with the scaled data, setting the original crypto names as the index
df_scaled = pd.DataFrame(df_market_data, columns=df_market_data.columns, index=crypto_names)

# Display sample data
df_scaled.head()
```

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Display the k-values
k_values
```

```python
# Create an empty list to store the inertia values
inertia_values = []

df_market_data_scaled = df_scaled

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for k in range(1, 12):

# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
# 2. Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(df_market_data_scaled)
# 3. Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans.inertia_)

# Display the inertia values
inertia_values
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": list(range(1, 12)),
    "inertia": inertia_values
}
elbow_data
# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

# Display the DataFrame
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = elbow_df.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Curve',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    width=600,
    height=400,
    line_width=2,
    grid=True
)

# Display the plot
elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** the plot shows a noticeable elbow at k=3

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
best_k = 3

# Initialize the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=best_k, n_init=10, random_state=42)

```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_clusters = kmeans_model.predict(df_market_data_scaled)
# Print the resulting array of cluster values.
crypto_clusters_df = pd.DataFrame(crypto_clusters, columns=['Cluster'])

# Print the first few rows of the resulting array of cluster values
print(crypto_clusters_df.head())
```

```python

# Create a copy of the DataFrame
df_market_data_copy = df_market_data_scaled.copy()



# Display the first few rows of the copied DataFrame to verify
df_market_data_copy.head()
```

```python
# Add a new column to the DataFrame with the predicted clusters
# Add the cluster labels to the copied DataFrame
df_market_data_copy['Cluster'] = crypto_clusters

# Display sample data
df_market_data_copy
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_transformed_data = pca.fit_transform(df_market_data_scaled)

# View the first five rows of the DataFrame. 
pca_transformed_data[ :5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance_ratio = pca.explained_variance_ratio_
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:**  0.9232885676615135 

```python
# Create a new DataFrame with the PCA data.

pca_df = pd.DataFrame(
    pca_transformed_data,
    columns=['PC1', 'PC2', 'PC3'],
    index=df_market_data.index  # Set the index to match the original data
)

# Copy the crypto names from the original data
pca_df['coin_id'] = df_market_data.index

# Set the coinid column as index
pca_df.set_index('coin_id', inplace=True)

# Display sample data
print(pca_df.head())
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
print(k_values)
```

```python
# Create an empty list to store the inertia values
inertia_values_pca = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
# Inside the loop:
    
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
# 2. Fit the model to the data using `df_market_data_pca`
    kmeans.fit(pca_df[['PC1', 'PC2', 'PC3']])
# 3. Append the model.inertia_ to the inertia list
    inertia_values_pca.append(kmeans.inertia_)

print(inertia_values_pca)   
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': list(range(1, 12)),
    'inertia': inertia_values_pca
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
elbow_df 
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = elbow_df.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Curve',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    width=600,
    height=400,
    line_width=2,
    grid=True
)

# Display the plot
elbow_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 3


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Doesn't differ much but the corresponding inertia values are lower.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=3, n_init=10, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
kmeans_model.fit(pca_df[['PC1', 'PC2', 'PC3']])
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto_clusters = kmeans_model.predict(pca_df[['PC1', 'PC2', 'PC3']])
# Print the resulting array of cluster values.
print(crypto_clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
df_pca_copy = pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
df_pca_copy['Cluster'] = crypto_clusters

# Display sample data
print(df_pca_copy.head())
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scatter_plot = df_pca_copy.hvplot.scatter(
    x='PC1',
    y='PC2',
    c='Cluster',         
    cmap='viridis',         
    hover_cols=['coin_id'], 
    title='Cryptocurrency Clusters PCA',
    xlabel='Principal Component 1 (PC1)',
    ylabel='Principal Component 2 (PC2)',
    width=600,
    height=400,
    size=10,
    grid=True
)

# Display the plot
scatter_plot
```
